## **Optimizacion no lineal convexa: caso inventarios**

Funcion no lineal cuadratica convexa de costo de inventario con 1 item
$$
z_i = b_i.D_i + k.\frac{D_i}{q_i} + \frac{1}{2}.q_i.i.b_i
$$

```python
# definimos la funcion de costo de inventario con 1 item
def costo_stock(q):
  return b*d +  k*d*(1/q) + 0.5*q*i*b
```

Ecuacion de lote optimo definida analiticamente para minimizar el costo de inventario. 
$$
q_{opt} = \sqrt{\frac{2.k.D}{b.P}}
$$

```python
# definimos la ecuacion de lote optimo obtenida analiticamente para minimizar el costo
def calcular_q_opt (k, d, b, i, ba = 0):
    return np.sqrt((2 * k * d)/(b * i + ba))
```

**Determinamos los parametros del ejercicio de inventarios**

```python

b = 5000 # costo por cada producto
k = 2000 # costo por cada orden de compra
d = 100 # demanda en el delta T
i = 0.01 # interes por almacenamiento
```

**Visualizamos la funcion de costo con 1 item**

```python
# Rango de q a visualizar
r_min, r_max = 20,300
# grilla de puntos para visualizar la funcion
x_range = np.arange(r_min, r_max, 5)
# computamos el costo para todos los valores del dominio q
costos_inventario_1item = costo_stock(x_range)
# create a line plot of input vs result
plt.plot(x_range, costos_inventario_1item)
plt.title('funcion Costo Inventarios')
plt.xlabel('variable de decision q')
plt.ylabel('funcion objetivo f(x)')
# show the plot
plt.show()
```

**Calculamos el lote optimo que minimiza el costo total usando la expresion analitica**

```python
calcular_q_opt (k, d, b, i, ba = 0)
```

**Calculamos el lote optimo que minimiza el costo total usando el metodo de gradiente descendiente**

$$
x_{n+1} = x_n - \gamma \bigtriangledown f(x)
$$

```python
# definimos la funcion de gradiente descendiente
def gradient_descent(gradient, x0, learn_rate, n_iter=50, tolerance=1e-06):
    # semilla
    xi = x0

    # vector vacio para ir guardando el valor de la variable de decision en cada paso 
    x_pasos = np.zeros(n_iter)
    # vector vacio para ir guardando el valor de la funcion objetivo en cada paso 
    func_pasos = np.zeros(n_iter)

    # guardo la variable de decision en el paso 1
    x_pasos[0] = xi

    # guardo la funcion objetivo con el xi actual
    func_pasos[0] = costo_stock(xi)

    # for loop para iterar paso-a-paso
    for i in range(n_iter):

        # diff calcula cuanto optimizo la funcion objetivo
        diff = -learn_rate * gradient(xi)

        # si la diff es menor a un umbral interrumpo la optimizacion
        if np.all(np.abs(diff) <= tolerance):
            break
        # actualizo la variable de decision actual con la mejora del ultimo paso    
        xi += diff

        # guardo la variable de decision en el paso 1 
        x_pasos[i] = xi
        # guardo la funcion objetivo con el xi actual
        func_pasos[i] = costo_stock(xi)

    return xi, x_pasos, func_pasos
```

```python
def grad_stocks_1item(x1):
    return 0.5*i*b-k*d*x1**(-2)
```

```python
x0 = 1000 
x_opt, x_pasos, func_pasos = gradient_descent(grad_stocks_1item, x0, learn_rate = 1, n_iter=50, tolerance=1e-06)
```
