## **Definimos funciones no lineales univariadas**

```python
def f1(x):
    return x**3 - 2*x**2 - 11*x + 12

def f2(x):
    return 2*x**2 - 3*x - 5
```

```python
#Initializing data
x = np.linspace(-4,6,100)
y1 = f1(x)
y2 = f2(x)
#y_dash = dfdx(x)


sns.set_context('talk', font_scale = 0.9)
#Plotting the function and its derivative
fig, ax = plt.subplots(1,2, figsize = (12,6))
ax[0].set_xlabel('x')
ax[1].set_xlabel('x')
ax[0].set_title('x**3 - 2*x**2 - 11*x + 12')
ax[1].set_title('2*x**2 - 3*x - 5')
#plt.axhline(0)
ax[0].plot(x,y1, label = 'f1(x)')
ax[1].plot(x,y2, label = 'f2(x)')
#plt.plot(x,y_dash, '--r', label = 'df(x)\dx')
ax[0].legend()
ax[1].legend()
plt.show()
```

```python
# objective function
def objective(x):
	return x**2.0

# define range for input
r_min, r_max = -5.0, 5.0
# sample input range uniformly at 0.1 increments
inputs = np.arange(r_min, r_max, 0.1)
# compute targets
results = objective(inputs)
# create a line plot of input vs result
plt.plot(inputs, results)
plt.title('funcion a optimizar')
plt.xlabel('variable de decision X')
plt.ylabel('funcion objetivo f(x)')
# show the plot
plt.show()
```

```python
# line plot of domain for a 1d function with optima and algorithm sample
from numpy import arange
from numpy.random import seed
from numpy.random import rand
from matplotlib import pyplot

# objective function
def objective(x):
	return x**2.0

# define range for input
r_min, r_max = -5.0, 5.0
# sample input range uniformly at 0.1 increments
inputs = arange(r_min, r_max, 0.1)
# compute targets
results = objective(inputs)
# simulate a sample made by an optimization algorithm
seed(1)
sample = r_min + rand(10) * (r_max - r_min)
# evaluate the sample
sample_eval = objective(sample)
# create a line plot of input vs result
plt.plot(inputs, results)
# define the known function optima
optima_x = 0.0
# draw a vertical line at the optimal input
plt.axvline(x=optima_x, ls='--', color='red')
# plot the sample as black circles
plt.plot(sample, sample_eval, 'o', color='black')
plt.title('funcion a optimizar')
plt.xlabel('variable de decision X')
plt.ylabel('funcion objetivo f(x)')
# show the plot
plt.show()
```

## **Funciones lineales bi-variadas**

```python
# filled contour plot for 2d objective function and show the optima
from numpy import arange
from numpy import meshgrid
from matplotlib import pyplot

# objective function
def f3(x, y):
	return x**2.0 + y**2.0

# define range for input
r_min, r_max = -5.0, 5.0
# sample input range uniformly at 0.1 increments
xaxis = arange(r_min, r_max, 0.1)
yaxis = arange(r_min, r_max, 0.1)
# create a mesh from the axis
x, y = meshgrid(xaxis, yaxis)
# compute targets
results = f3(x, y)
# create a filled contour plot with 50 levels and jet color scheme
plt.contourf(x, y, results, levels=50, cmap='jet')
# define the known function optima
optima_x = [0.0, 0.0]
# draw the function optima as a white star
plt.plot([optima_x[0]], [optima_x[1]], '*', color='white')
# show the plot
plt.colorbar()
plt.xlabel('Variable de decision X1')
plt.ylabel('Variable de decision X2')
plt.title('Curvas de iso-costo de la funcion objetivo basados en X1 y X2')
plt.show()
```

## **Funciones de inventario**

```python
def costo_stock(b, x1, k, d, i):
  return b*d +  k*d*(1/x1) + 0.5*x1*i*b
```

```python
b = 5000
k = 2000
d = 100
i = 0.01

# define range for input
r_min, r_max = 20,300
# sample input range uniformly at 0.1 increments
x_range = np.arange(r_min, r_max, 5)
# compute targets
results = costo_stock(b, x_range, k, d, i)
# create a line plot of input vs result
plt.plot(x_range, results)
plt.title('funcion a optimizar')
plt.xlabel('variable de decision X')
plt.ylabel('funcion objetivo f(x)')
# show the plot
plt.show()
```

```python
def costo_stock_2d(b, x1, x2, k, d, i):
  return b*d +  k*d*(1/x1) + 0.5*x1*i*b + b*d +  k*d*(1/x2) + 0.5*x2*i*b
```

```python


# define range for input
#r_min, r_max = 0, 400.0
# sample input range uniformly at 0.1 increments
xaxis = arange(20, 300, 5)
yaxis = arange(20, 300, 5)
# create a mesh from the axis
x1_range, x2_range= meshgrid(xaxis, yaxis)
# compute targets
results = costo_stock_2d(b, x1_range, x2_range, k, d, i)
# create a filled contour plot with 50 levels and jet color scheme
plt.figure(figsize=(8, 6), dpi=80)
plt.contourf(x1_range, x2_range, results, levels=40, cmap='plasma',)
# define the known function optima
#ptima_x = [0.0, 0.0]
# draw the function optima as a white star
#plt.plot([optima_x[0]], [optima_x[1]], '*', color='white')
# show the plot
plt.colorbar()
plt.xlabel('Variable de decision X1')
plt.ylabel('Variable de decision X2')
plt.title('Funcion stock con 2 items X1 y X2')
plt.show()
```
